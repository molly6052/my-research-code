# トリミング関数　定義

```python
import os
from PIL import Image
import numpy as np
from PIL import ImageFile
from keras.preprocessing.image import array_to_img
```

```python
#入力：画像,トリミングサイズ縦(km)、トリミングサイズ横(km)
def crop_center(img, crop_width, crop_height):
    crop_width = (crop_width*512)/2600
    crop_height = (crop_height*512)/2600
    img_width, img_height, channels = img.shape
    return img[int((img_height - crop_height) / 2):int((img_height + crop_height) / 2),
            int((img_width - crop_width) / 2):int((img_width + crop_width) / 2)]
```

# Auto Encoderモデルを準備する

###### TensorFlowおよびその他のライブラリをインポートする

```python
import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
X_train = np.reshape(x_train, (len(x_train), 112, 112, 3))
X_test = np.reshape(x_test, (len(x_test), 112, 112, 3))
X_valid = np.reshape(x_valid, (len(x_valid), 112, 112, 3))
print(X_train.shape)
print(X_test.shape)
print(X_valid.shape)
```

```python
img_shape = X_train.shape[1:]
img_shape
```

```python
# Create the Encoder and Decoder
#pass the gray scale input image of size(112,112,3)
inputs = tf.keras.Input(shape=img_shape, name='input_layer')
# Conv Block 1 -> BatchNorm->leaky Relu
encoded = tf.keras.layers.Conv2D(16, kernel_size=3, strides= 1, padding='same', name='conv_1')(inputs)
encoded = tf.keras.layers.BatchNormalization(name='batchnorm_1')(encoded)
encoded = tf.keras.layers.LeakyReLU(name='leaky_relu_1')(encoded)
# Conv Block 2 -> BatchNorm->leaky Relu
encoded = tf.keras.layers.Conv2D(32, kernel_size=3, strides= 2, padding='same', name='conv_2')(encoded)
encoded = tf.keras.layers.BatchNormalization(name='batchnorm_2')(encoded)
encoded = tf.keras.layers.LeakyReLU(name='leaky_relu_2')(encoded)
# Conv Block 3 -> BatchNorm->leaky Relu
encoded = tf.keras.layers.Conv2D(64, kernel_size=3, strides=2, padding='same', name='conv_3')(encoded)
encoded = tf.keras.layers.BatchNormalization(name='batchnorm_3')(encoded)
encoded = tf.keras.layers.LeakyReLU(name='leaky_relu_3')(encoded)
#Decoder
# DeConv Block 1-> BatchNorm->leaky Relu
decoded = tf.keras.layers.Conv2DTranspose(64, 3, strides= 1, padding='same',name='conv_transpose_1')(encoded)
decoded = tf.keras.layers.BatchNormalization(name='batchnorm_4')(decoded)
decoded = tf.keras.layers.LeakyReLU(name='leaky_relu_4')(decoded)
# DeConv Block 2-> BatchNorm->leaky Relu
decoded = tf.keras.layers.Conv2DTranspose(32, 3, strides= 2, padding='same', name='conv_transpose_2')(decoded)
decoded = tf.keras.layers.BatchNormalization(name='batchnorm_5')(decoded)
decoded = tf.keras.layers.LeakyReLU(name='leaky_relu_5')(decoded)
# DeConv Block 3-> BatchNorm->leaky Relu
decoded = tf.keras.layers.Conv2DTranspose(16, 3, 2, padding='same', name='conv_transpose_3')(decoded)
decoded = tf.keras.layers.BatchNormalization(name='batchnorm_6')(decoded)
decoded = tf.keras.layers.LeakyReLU(name='leaky_relu_6')(decoded)
# output
outputs = tf.keras.layers.Conv2DTranspose(3, 3, 1,padding='same', activation='sigmoid', name='conv_transpose_4')(decoded)
```

```python
outputs.shape
```

```python
def SSIMLoss(y_true, y_pred):
  return 1 - tf.reduce_mean(tf.image.ssim(y_true, y_pred,1.0))
```

```python
autoencoder = tf.keras.Model(inputs, outputs)
optimizer = tf.keras.optimizers.Adam(learning_rate = 0.0005)
autoencoder.compile(optimizer=optimizer, loss=SSIMLoss)
```

```python
autoencoder.summary()
```

```python
#　層を図で表示する
from tensorflow.keras.utils import plot_model

plot_model(autoencoder, to_file="architecture.png", show_shapes=True)
```

```python
history = autoencoder.fit(X_train, X_train,
                epochs=400,
                batch_size=6,
                shuffle=True,
                validation_data=(X_valid[10:20], X_valid[10:20])
                )
```

# 検証データの通常発達をデコード後と比較

```python
decoded_imgs = autoencoder.predict(X_valid)
classes = ['RI', 'NORMAL']
n = 15
plt.figure(figsize=(30, 4))
for i in range(11, n + 1):
    # Display original
    ax = plt.subplot(2, n, i)
    plt.imshow(((X_valid[i]*255.).astype(np.uint8)))
    plt.gray()
    ax.get_xaxis().set_visible(False)
    plt.title("Original:  " + str(classes[np.argmax(y_valid[i], axis=0)]), fontsize = 8)
    ax.get_yaxis().set_visible(False)
    # Display reconstruction
    ax = plt.subplot(2, n, i + n)
    plt.title("Reconstructed:  " + str(classes[np.argmax(y_valid[i], axis=0)]), fontsize = 8)
    plt.imshow(((decoded_imgs[i]*255.).astype(np.uint8)))
    plt.gray()
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
plt.show()
```

```python
n = 6  # how many encoded and decoded images we will display
decoded_imgs= autoencoder.predict(X_valid)
decoded_imgs1= autoencoder.predict(X_train)
decoded_anamolus= autoencoder.predict(X_test)
plt.figure(figsize=(20, 14), dpi=100)
plt.subplots_adjust( wspace=0.1, hspace=0.07)
plt_a=1
for i in range(n):
    # Original training dataset vs Original training
    ax = plt.subplot(3, n, plt_a   )
    plt.imshow((X_train[i]*255.).astype(np.uint8))
    ax.get_xaxis().set_visible(True)
    ax.get_yaxis().set_visible(False)
    value_a = SSIMLoss(decoded_imgs1[i], X_train[i])
    ax.set_title("Original Image : " + str(classes[np.argmax(y_train[i], axis=0)]), fontsize = 10)
    label = 'SSIM Loss value: {:.3f}'
    ax.set_xlabel(label.format(value_a) )
    
    # Reconstructed good data  vs Original training data
    ax = plt.subplot(3, n, plt_a + n )
    plt.imshow((decoded_imgs[i]*255.).astype(np.uint8))
    ax.get_xaxis().set_visible(True)
    ax.get_yaxis().set_visible(False)    
    value_a = SSIMLoss(decoded_imgs[i], X_valid[i])
    ax.set_title("Reconstructed Image : " + str(classes[np.argmax(y_valid[i], axis=0)]), fontsize = 10)
    label = 'SSIM Loss value: {:.3f}'
    ax.set_xlabel(label.format(value_a) )
    
    # Reconstructed anomalous data  vs Original training data
    ax = plt.subplot(3, n, plt_a + 2*n)
    plt.imshow((decoded_anamolus[i]*255.).astype(np.uint8))
    ax.get_xaxis().set_visible(True)
    ax.get_yaxis().set_visible(False)
    value = SSIMLoss(decoded_anamolus[i], X_test[i])
    label = 'SSIM Loss value: {:.3f}'
    ax.set_title("Anamolus Image : " + str(classes[np.argmax(y_test[i], axis=0)]), fontsize = 10)
    ax.set_xlabel(label.format(value) )
    plt_a+=1
plt.show()
```

# Findthreshold

```python
def find_threshold(model, x_train_scaled):
  ssim = []
  reconstructions = model.predict(x_train_scaled)
  for i in range(len(x_train_scaled)):
    
    # provides losses of individual instances
    reconstruction_errors = SSIMLoss(reconstructions[i], x_train_scaled[i]).numpy()
    ssim.append(reconstruction_errors)
    
  # threshold for anomaly scores
  threshold = np.mean(ssim) \
      + np.std(ssim)
  return threshold

def find_threshold_method_two(model, x_train_scaled):
  # another method to find threshold
  ssim2 = []
  reconstructions = model.predict(x_train_scaled)
  # provides losses of individual instances
  reconstruction_errors = SSIMLoss(reconstructions[i], x_train_scaled[i]).numpy()

  threshold_2 = np.percentile(reconstruction_errors, 95)
  return threshold_2

def get_predictions(model, x_test_scaled, threshold):
  ssimloss = []
  preds = []
  predictions = model.predict(x_test_scaled)
  for i in range(len(x_test_scaled)):
    # provides losses of individual instances
    errors = SSIMLoss(predictions[i], x_test_scaled[i]).numpy()
    ssimloss.append(errors)
    # 0 = anomaly, 1 = normal
    anomaly_mask = pd.Series(ssimloss[i]) > threshold

    pred = anomaly_mask.map(lambda x: 0.0 if x == True else 1.0)
    preds.append(pred)
  return np.array(preds),ssimloss
```

```python
threshold = find_threshold(autoencoder, X_train)
print(f"Threshold method one: {threshold}")

threshold_2 = find_threshold_method_two(autoencoder, X_train)
print(f"Threshold method two: {threshold_2}")
```
